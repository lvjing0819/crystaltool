```python
import pandas as pd
file_path = r"/Volumes/Elements/跨境电商/TIKTOK/账务/全部 订单-2024-10-03.xlsx"
df = pd.read_excel(file_path)
```

```python
df.shape
df=df.drop(index=0)
df
```

```python
"""指标	口径解释
已结算-成功付款	
Total settlement amount！= Nan&Cancelation/Return Type=空白&Total settlement amount>=0
已结算-派送问题付快递费（拒收/3次未送达退货商家）	
Total settlement amount！= Nan&Cancelation/Return Type=cancel&The order is cancelled by whom=system
已结算-达人样品	
Total settlement amount！= Nan&Cancelation/Return Type=空白&Total settlement amount<0
已结算-客户退款	
Total settlement amount！= Nan&Cancelation/Return Type=refund
已结算-取消订单	
Total settlement amount！= Nan&Cancelation/Return Type=cancel&The order is cancelled by whom!=system
未结算-取消（顾客取消，商家取消，系统取消-超时揽收，超时支付，派送问腿未退回商家）	
Total settlement amount= Nan&Current order status=canceled&completed
未结算-未揽收+派送中	
Total settlement amount= Nan&(Current order status=To ship|（Current order status=Shipped&Order Substatus=In Transit）)
未结算-已送达未达结算日期	
Total settlement amount= Nan&Current order status=Shipped&Order Substatus=Delivered
未结算-未支付（未达取消日期）	
Total settlement amount= Nan&Current Order Status=Unpaid
"""
```

```python
#已结算-成功付款	
#Total settlement amount！= Nan&Cancelation/Return Type=空白&Total settlement amount>=0
	
df_settled_success=df[(df["Total settlement amount"]>0.0000001) & (pd.isna(df['Cancelation/Return Type']))]
df_settled_success.shape
# >=0:3135 rows × 105 columns

#df[df["Total settlement amount"]>=0]["Total settlement amount"]
#df[pd.isna(df['Cancelation/Return Type'])]["Cancelation/Return Type"]

```

```python
#已结算-成功付款
df_settled_success['SKU ID'].value_counts()
df_settled_success.groupby('SKU ID')['Total settlement amount'].sum()
```

```python
#已结算-派送问题付快递费（拒收/3次未送达退货商家）
#Total settlement amount！= Nan&Cancelation/Return Type=cancel&The order is cancelled by whom=system
df_settled_dilivery_error=df[(pd.notna(df['Total settlement amount'])) & (df['Cancelation/Return Type']=="Cancel") &(df['Cancel By']=="System")]
df_settled_dilivery_error['Total settlement amount']
df_settled_dilivery_error.shape
# 47
```

```python
#已结算-达人样品	
#Total settlement amount！= Nan&Cancelation/Return Type=空白&Total settlement amount<0

df_settled_influncer_sample=df[(pd.notna(df['Total settlement amount'])) & (pd.isna(df['Cancelation/Return Type'])) &(df['Total settlement amount'] <0)]
df_settled_influncer_sample.shape
# 
```

```python
#已结算-客户退款	
#Total settlement amount！= Nan&Cancelation/Return Type=refund
df_settled_customer_refund=df[(pd.notna(df['Total settlement amount'])) & (df['Cancelation/Return Type']== "Return/Refund")]
df_settled_customer_refund.shape
```

```python
#已结算-取消订单
#Total settlement amount！= Nan&Cancelation/Return Type=cancel&The order is cancelled by whom!=system
df_settled_customer_cancel=df[(pd.notna(df['Total settlement amount'])) & (df['Cancelation/Return Type']== "Cancel")&(df['Cancel By']!="System")]
df_settled_customer_cancel.shape
```

```python
#未结算-取消（顾客取消，商家取消，系统取消-超时揽收，超时支付，派送问腿未退回商家）	
#Total settlement amount= Nan&Order Status=canceled or completed
df_unsettled_canceled=df[(pd.isna(df['Total settlement amount'])) & ((df['Order Status']== "Canceled")|(df['Order Status']== "Completed"))]
df_unsettled_canceled.shape
```

```python
#未结算-未揽收+派送中	
#Total settlement amount= Nan&(Current order status=To ship|（Current order status=Shipped&Order Substatus=In Transit）)
df_unsettled_on_delivery=df[(pd.isna(df['Total settlement amount'])) & ((df['Order Status']== "To ship")|((df['Order Status']== "Shipped")&(df['Order Substatus']== "In Transit")))]
df_unsettled_on_delivery.shape
```

```python
#未结算-已送达未达结算日期	
#Total settlement amount= Nan&Current order status=Shipped&Order Substatus=Delivered
df_unsettled_date_not_reached =df[(pd.isna(df['Total settlement amount'])) & ((df['Order Status']== "Shipped")&(df['Order Substatus']== "Delivered"))]
df_unsettled_date_not_reached.shape
```

```python
#未结算-未支付（未达取消日期）	
#Total settlement amount= Nan&Current Order Status=Unpaid
df_unsettled_unpaid=df[(pd.isna(df['Total settlement amount'])) & (df['Order Substatus']== "Unpaid")]
df_unsettled_unpaid.shape
```
